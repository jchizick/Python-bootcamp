```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Read data from file
df = pd.read_csv('epa-sea-level.csv')
df.head
```

```python
df.columns
```

```python
#--

fig, ax = plt.subplots(figsize = (12,4))
ax.scatter(df['Year'], df['CSIRO Adjusted Sea Level'], color='red',alpha=0.3)
ax.set(xlabel="Year", ylabel="CSIRO Adjusted Sea Level",title="Rise in Sea Level")

#--
stats = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
m = stats.slope
b = stats.intercept
x_line = range(1880, 2050)

plt.plot(x_line, m * x_line + b, color='black',linestyle="dashed")

#--
df_2000 = df[df['Year'] >= 2000]
stats = linregress(df_2000['Year'], df_2000['CSIRO Adjusted Sea Level'])
m = stats.slope
b = stats.intercept
x_line = range(1950, 2050)

plt.plot(x_line, m * x_line + b,color="black",linestyle="--")

```
